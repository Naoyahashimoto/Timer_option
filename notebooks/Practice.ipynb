株価のSDEに従うサンプルパスの生成&描画

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_path(S0, T, r, vol, N, M):
    dt = T/M
    w = np.cumsum(np.reshape(np.random.standard_normal(N*M), (N,M)), 1) * (dt**0.5)
    t = np.cumsum(np.ones((N,M)), 1)*dt   
    return S0 * np.exp((r-0.5*vol**2)*t + vol*w)

if __name__ == "__main__":
    
    S0  = 100.0
    T   = 3.0
    R   = 0.01
    VOL = 0.2
    
    N = 50
    M = 200
    x = generate_path(S0, T, R, VOL, N, M)
    plt.plot(x.T)
    plt.show()
```

モンテカルロ

```python
import numpy as np
import matplotlib.pyplot as plt
S0 = 60.0
r = 0.02
sigma = 2
T = 1.0 #1 Time cycle
N = 100 #steps
dt = T / N #dt

def genS_py(I):
    S = []
    for i in range(I):
        path = []
    for t in range (N + 1):
        if t == 0:
            path.append(S0)
        else:
            z = normal(0.0, 1.0)
            St = path[t - 1]*np.exp((r - 0.5*sigma ** 2)*dt + sigma*np.sqrt(dt)*z)
            path.append(St)
            S.append(path)
            return S
I = 100
%time S = genS_py(I)
```
